###Creating Sample Data Frame

```python
data = [("iPhone", 999), 
        ("Samsung", 799), 
        ("MacBook", 1299),
        ("Vivo", 499)]
```

```python
df = spark.createDataFrame(data, ["product", "price"])
display(df)
```

### Doing Filter on the DF###

```python
df.filter(df.price > 1000).show()
```

###Getting Total Price

```python
from pyspark.sql.functions import *

df.select(sum("price")).show()
```

###Getting Total Avg Price

```python
df.select(avg("price")).show()
```

### Getting Product that is having price higher then Average Price

```python
df.filter(df.price > df.select(avg("price")).first()[0]).show()
```

###Adding Boolean Column base on price greter or less then avg price

```python
df = df.withColumn("is_abv_avg", col("price") > df.select(avg("price")).first()[0])
```

```python
df.show()
```

###Sorting Column base on the desceding Price

```python
df.orderBy(col("price").desc()).show()
```

```python
df.filter(df.price == df.select(max("price")).first()[0]).show()
```
